```python

# nombre_carpeta = 'images_pred'
nombre_carpeta = 'prueba2'

direccion2 = 'C:/Users/guavitaj/Documents/tesis/algoritmos/modelo_xray'

ruta_carpeta = os.path.join(direccion2,nombre_carpeta)

if not os.path.exists(ruta_carpeta):
    os.makedirs(ruta_carpeta)
    print(f'Carpeta creada: {ruta_carpeta}')
else:
    print(f'La carpeta ya existe: {ruta_carpeta}')


fecha_inicio = datetime.strptime("202308010000", "%Y%m%d%H%M")
fecha_fin = datetime.strptime("202309010000", "%Y%m%d%H%M")

fecha_str = os.path.basename(ncs[2]).split('_')[3][1:13]
print(fecha_str)
fecha_archivo = datetime.strptime(fecha_str, "%Y%m%d%H%M")
print(fecha_archivo)
```

```python
for nc in ncs:
    #ds = xr.open_dataset(nc).sel(x=slice(-0.01, 0.03), y=slice(0.05, 0.01))

    fecha_str = os.path.basename(nc).split('_')[3][1:13]
    fecha_archivo = datetime.strptime(fecha_str, "%Y%m%d%H%M")

    if fecha_inicio <= fecha_archivo <= fecha_fin:

        print("fecha archivo:",fecha_archivo)

        ds = xr.open_dataset(nc)
        variable = ds['ACM']
        variable.plot()

        output_path = os.path.join(ruta_carpeta, f"{os.path.basename(nc).replace('.nc', '.png')}")
        plt.savefig(output_path)
        
        # Cerrar la figura para liberar memoria
        plt.close()
        print(f'Imagen guardada para el archivo: {nc}')

    else:

        print(f'Archivo {nc} fuera del rango de fechas.')
```
